```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
sns.set()
```

```python
elnino = pd.read_csv('elnino.csv')
elnino = elnino.replace('.',0)
elnino.head(10)
```

```python
husein = pd.read_csv('TESLA.csv')
husein.head()
```

```python
husein.Close.shape
```

```python
husein_elnino = elnino.iloc[-husein.shape[0]:, 7:]
husein_elnino.reset_index(drop=True, inplace=True)
```

```python
husein = pd.concat([husein[['Date', 'Close']], husein_elnino], axis=1, ignore_index=True)
husein.columns = ['date','catch','Zonal Winds', 'Meridional Winds', 'Humidity', 'Air Temp', 'Sea Surface Temp']
husein.head()
```

```python
husein_scaled = pd.DataFrame(MinMaxScaler().fit_transform(husein.iloc[:, 1:]))
husein_scaled.columns = ['catch','Zonal Winds', 'Meridional Winds', 'Humidity', 'Air Temp', 'Sea Surface Temp']
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(husein_scaled.corr(), annot=True)
plt.show()
```

```python
husein_scaled.to_csv('husein_scaled.csv',index=False)
husein.to_csv('husein.csv',index=False)
```
